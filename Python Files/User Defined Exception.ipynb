## Custom Exception:


### Creating custom exception

```python
"""
-> we can create new custom exception by creating new class.
-> exception class must have derived either directly or indirectly from built-in exception class.
-> Most of the built in exception derived from this class.
-> every exception in program are placed in exception.py or error.py files
"""
class CustomError(Exception):
    pass
raise CustomError("An Error Occured")
```

### User defined exception in Python

```python
"""
-> used to raise and catch the error
##guessing game
"""
class Error(Exception):
    """Base class for other exceptions"""
    pass

class ValueTooSmallError(Error):
    """Raised when the input value is too small"""
    pass

class ValueTooLargeError(Error):
    """Raised when the input value is too large"""
    pass

number = 22 #guessing number

while True:
    try:
        input_num = int(input("Enter a number: "))
        if input_num < number:
            raise ValueTooSmallError
        elif input_num > number:
            raise ValueTooLargeError
        break
    except ValueTooSmallError:
        print("This value is too small, try again.")
        print()
    except ValueTooLargeError:
        print("This value is too large, try again.")
        print()
print("Congratulations!, you guessed it correctly.")
               
```

### Customizing Exception classes

```python
"""
-> can customize exception classes to accept other arguments as per our needs.
"""
class SalaryNoInRangeError(Exception):
    """Exception raised for errors in input salary
    
    Attributes:
        salary-- input salary which caused error
        message-- explanation of the error 
    """
    def __init__(self, salary, message="Salary not in (5000, 15000) range"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)

salary = int(input("Enter salary amount: "))
if not 5000 < salary < 15000:
    raise SalaryNoInRangeError(salary)
```
