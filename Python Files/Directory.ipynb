## Directories

```python
"""
-> if there are number of files we can arrane our code within different directories to make more manageble
-> directories is collection of files and subdirectories
-> 'os' module which provide many useful methods to work with directories(and file as well)
"""
```

### Get current directory

```python
"""
-> we can get present working directory using getcwd() method of 'os' module
-> returns current working directory in string form
-> we can use getcwdb() to get it as bytes object
"""
import os
print(os.getcwd())

print(os.getcwdb())

"""
-> Extra blckslash implies an escape sequence
-> print() will render this properly
"""
```

### List Directories and files

```python
"""
-> all files and sub-directories inside directory can be retrieved using listdir() method
-> takes a path and returns a list of subdirectories and files in that path
-> if no path is specified, returns list of subdirectories and files from current working directories
"""
print(os.getcwd())
os.listdir()
```

### making a new directory

```python
"""
-> we can make new directory using mkdir() method
-> takes path of new directory
-> if full path is not specified, new directory is created in current working directory
"""
#os.mkdir("testing")
os.listdir()
```
