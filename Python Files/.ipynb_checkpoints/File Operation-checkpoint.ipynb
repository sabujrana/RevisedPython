## File:

```python
"""
-> Files are name locations on disk to stored related information.
-> used to permanently store data in non-volatile memory(hard disk)
-> Since Random Access Memory (RAM) is volatile (which loses its data when the computer is turned off), 
we use files for future use of the data by permanently storing them.

-> When we want to read from or write to a file, we need to open it first.

-> When we are done, it needs to be closed so that the resources that are tied with the file are freed.

File Operation in Python:
(1) Open a file
(2) Read or write(perfom operation)
(3) Close file

"""
```

### Opening File

```python
"""
-> used built-in open() function is used to open file
-> return file object; called handle, used to read or modify file accordingly
"""
#opening file in current directory
f = open("hello.txt") 
#specifying full path
#f = open("/home/sabuz/Documents/Techdocs/Aago/Final Revision of Python/")

"""
-> we can specify mode while opening file
-> In mode we specify, whether we want read r, write w or append a to file
-> we can also specify if we want to open file in text mode or binary mode
-> default is reading in text mode, in this we get strings when reading from the file
-> binary mode return bytes and this is used when daling with non-text file like images or executable

Mode:r -read
     w -write, truncate if file exist, create if does not exist
     a -append, at the end without truncate, create new one if does not exist 
     x -exclusive creation, file alerady exist operation failed
     t -text mode
     b -binary
     + -open files for updating(reading and writing)
"""
f = open("hello.txt")# equivalent to 'r' or 'rt'
f = open("hello.txt", 'w')#write mode
f = open("hello.txt",'r+b') #read and write in binary mode
"""
-> unlike other Pl 'a' does not imply numbeer 97 until encoding in ASCII(or other equivalent encoding)
-> default encoding is platform dependent i.e for windows cp1252 but utf-8 for linux
-> we don't rely on default encoding or else our code will behave diffrent in different platform
-> It is highly recommend to specify encoding type while working with file

"""
f = open("hello.txt", mode='r', encoding='utf-8')

```

### Closing Files

```python
"""
-> After completing operation on file we need to properly close file.
-> closing file will free up resources that are tied with file, which is done using close() method 
-> has garbage collector to clean up unreferenced object but we mustn't rely on it to close file
"""
f = open("hello.txt", encoding="utf-8")
# perform file operation
f.close()

"""
-> this method is not entirely safe, if expception occurs while perfroming operation, code exits without closing
file. So safe method is to use try...finally block
"""
try:
    f = open("hello.txt", encoding='utf-8')
    #file operation
finally:
    f.close()
    
"""
-> this way guarenteed closing of file even if exception is occured that causes program flow to stop
-> best way to close file is by using 'with' statement; which ensures that file is closed when block inside with 
statemnt is exited.
-> we don't need to explicitly call close() method; will done internally
"""

with open("hello.txt", encoding='utf-8') as f:
    #file operation
```

### Writing to file:

```python
"""
-> To write in a file we shoud open it in 'w', append 'a' or exclusive creation 'x' mode
-> 'w' will overwrite file as it already exist so we must be careful, which will erased previous data
-> writing string or sequence of byte(binary files) are done by using write() method; returns no of character
written in file
"""
with open("example.txt", 'w', encoding="utf-8") as f:
    f.write("This is first line.\n")
    f.write("This is second line.\n")
    f.write("End of file.\n")
    
"""
-> will create new textfile in current directory if does not exist, overwrite if already exist
-> must include newline character ourselves to distinguish different lines.
"""
```

### Reading Files

### File Methods

```python
"""
close(): closes open file, no effect if file is already closed
detach(): separates underlying binary buffer from 'TextIOBase' and return it
fileno(): returns interger number(file descriptor) of file
flush(): flushes write buffer of file stream
isatty(): return 'True' if file stream is interactive
read(n): read most n character from file, reads till end of file if it is negative or 'None'
readable(): returns 'True' if file stream can be read from
readline(n=-1): reads and return one line from file, reads at most 'n' bytes if specified
seek(offset,from=SEEK_SET): changes the file position to 'offset' bytes, in reference to from(start,current,end)
seekable(): return 'True' if file stream supports random access
tell(): returns current file location
truncate(size=None): resizes the file stream to 'size' bytes, if 'size' is not specified, resizes to current location
writable(): return 'True' if file stream can be written to
write(s): write string 's' to file and return the number of charaters written
writelines(lines): write a list of 'lines' to the file
"""
```

```python
with open('hello.txt','w', encoding='utf-8') as f:
    f.writelines("I also want to learn JS.")
    
```
