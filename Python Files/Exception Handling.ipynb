## Exception 

```python
"""
-> There are many built-in exception raised when program encounters an error(sth that program goes wrong)
-> When exception occurs interpreter stops current process and passes it to calling process until it handled.
-> If exception is not handled program will crash.
-> If we have function A calls B and it calls C, if exception occurs in C and not handled; exception passes to B 
and A
-> If exception never handled, error message is displayed and our program comes to sudden unexpected halt
"""
```

### Catching Exception

```python
"""
-> exception is handled using 'try' statement
-> critical statement that can raise exception is placed inside of 'try' block
-> The code which handles exception is written in 'except' clause
-> Thus we can choose what operation to perform once we have caught exception
"""
import sys

randomList = ['i', 0, 2]

for entry in randomList:
    try:
        print(f"The entry is: {entry}.")
        r = 1/int(entry)
        break
    except:
        print("Oops!",sys.exc_info()[0],"occured.")
        print("Next Entry")
        print()
print("The reciprocal of", entry, "is", r)

"""
-> we placed the block of code which raise exception inside of the 'try' block
-> If no exception occured, 'except' block is skipped and normal flow continues till last values
-> If exception occured, it is caught by 'except' block
-> We print exception using exec_info() function inside 'sys' module
"""
```

```python
"""
-> Every exception in python inherits from base 'Exception' class
"""
import sys

randomList = ['h', 0, 3]

for entry in randomList:
    try:
        print(f"The entry is {entry}.")
        r = 1/int(entry)
        break
    except Exception as e:
        print("Oops!",e.__class__, "occured.")
        print("Next Entry.")
        print()
        
print("The reciprocal of", entry, "is", r)
```

### Raising Exception:

```python
"""
-> exception are raised when error is occured at runtime
-> we can manually raise exception using 'raise' keyword
-> we can optionally pass values to exception to clarify why that exception is raised

"""
try:
    a = int(input("Enter a value: "))
    if a <= 0:
        raise ValueError("That is negative number.")
except ValueError as ve:
    print(ve)
```

### try with else clause:

```python
#assert keyword
"""
-> mainly used for debugging
-> returns nothing if condition of asset is True, either AssertionError will raised
-> can also provide error messege if condition is False
"""
name = "sabuj"
assert name == 'Sabuj',"name not matched"
```

```python
"""
-> If we want to run certain block of code,if code inside 'try' block ran without error we can use optional 'else'
Note: exception in 'else' clause are not handled by the preceding except clause
"""
#to print reciprocal of even numbers
try: 
    number = int(input("Enter a number: "))
    assert number % 2 == 0
except: 
    print("Not even number.")
else:
    reciprocal = 1/number
    print(reciprocal)
    
"""
->if we pass even number,reciprocal will return;odd number returns No even number and 0 returns "ZeroDivisionError"
"""
```
